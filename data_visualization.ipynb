# Data Visualization

<li>Data Visualization is the process of presenting data in the form of graphs or charts.</li>
<li>Data visualisation is an essential part of any data analysis or machine learning workflow. It allows to gain insights about our data.</li>
<li>As the famous saying suggests, 'a picture is worth a thousand words'.</li>
<li>It helps to understand large and complex amounts of data very easily.</li>
<li>It allows the decision-makers to make decisions very efficiently and also allows them in identifying new trends and patterns very easily.</li>
<li>It is also used in high-level data analysis for Machine Learning and Exploratory Data Analysis (EDA).</li>
<li>At the heart of any data science workflow is data exploration. Most commonly, we explore data by using the following:</li>
<ol>
    <li>Statistical methods (measuring averages, measuring variability, etc.)</li>
    <li>Data visualization (transforming data into a visual form)</li>
</ol>

<li>This indicates that one of the central tasks of data visualization is to help us explore data.</li>
<li>The other central task is to help us communicate and explain the results we've found through exploring data.</li>
<li>That being said, we have two kinds of data visualization:</li>
<ol>
    <li><b>Exploratory data visualization:</b> we build graphs for ourselves to explore data and find patterns.</li>               <li><b>Explanatory data visualization:</b> we build graphs for others to communicate and explain the patterns we've found through exploring data.</li>
</ol>
<li>Data visualization can be done with various tools like Tableau, Power BI, Matplotlib, Plotly, Seaborn.</li>
<li>Here, we are going to learn about matplotlib and seaborn.</li>

## Coordinate Systems In A Graph

<li>We can create a graph by drawing two lines at right angles to each other.</li>
<li>Each line is called an axis — the horizontal line at the bottom is the x-axis, and the vertical line on the left is the y-axis.</li>
<li>The point where the two lines intersect is called the origin.</li>

![](images/axis_and_origin.png)

<li>Each axis has length — below, we see both axes marked with numbers, which represent unit lengths.</li>

![](images/unit_distance_from_axis.png)

<li>The length of the axes helps us precisely locate any point drawn on the graph.</li>
<li>Point A on the graph below, for instance, is seven length units away from the y-axis and two units away from the x-axis.</li>
<li>The x-coordinate shows the distance in unit lengths relative to the y-axis.</li>
<li>The y-coordinate shows the distance in unit lengths relative to the x-axis.</li>

### Growth over time and thier graphs
<li>Generally, a quantity that increases very quickly in the beginning — and then it slows down more and more over time — has a logarithmic growth.</li>
<li>Generally, a quantity that increases slowly in the beginning — but then starts growing faster and faster over time — has exponential growth.</li>
<li>Generally, a quantity that increases constantly over time has linear growth.</li>
<li>To sum up, these are the three types of growth we've learned in this session:</li>

![](images/types_of_graph.png)

```python
import numpy as np
```

```python
df = pd.DataFrame({'original': [i for i in range(1, 11)],
                  'linear': [i+5 for i in range(1, 11)],
                  'logarithmic':[np.log(i) for i in range(1, 11)],
                  'exponential': [i**i for i in range(1, 11)]})
df.head()
```

```python
plt.figure(figsize=(8,6))
plt.plot(df['original'], df['linear'])
plt.title('Linear Graph')
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.plot(df['original'], df['logarithmic'])
plt.title('Logarithmic Graph')
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.plot(df['original'], df['exponential'], marker='o')
plt.title('Exponential Graph')
plt.show()
```

## Distribution Plots

<li>It's useful to examine the shape of a histogram because it shows us the distribution of the values.</li>
<li>We often see histograms with a shape that is more or less symmetrical.</li>
<li>If we draw a vertical line exactly in the middle of a symmetrical histogram, then we divide the histogram in two halves that are mirror images of one another.</li>
<li>A histogram shows the distribution of the values, and if its shape is symmetrical, then we say we have a symmetrical distribution.</li>
<li>One common symmetrical distribution is the normal distribution (also called Gaussian distribution).</li>


![](images/symmetric_distribution.png)

```python
import numpy as np
a = np.random.normal(loc = 20, scale = 1, size = 100)
```

```python
sns.distplot(a)
```

### Normal Distribution
<li>A normal distribution is a type of continuous probability distribution in which most data points cluster toward the middle of the range.</li>
<li>While the rest taper off symmetrically toward either extreme. The middle of the range is also known as the mean of the distribution.</li>
<li>Most of the values in that column lie in the middle of the range.</li>
<li>As we approach the minimum and the maximum points of the range, we have less and less of those values.</li>

![](images/normal_dist2.png)

### Uniform Distribution

<li>Another symmetrical distribution we can see in practice is the uniform distribution</li>
<li>The values are uniformly distributed. The bars have equal height because the intervals have equal frequencies.</li>
<li>When we plot histograms in practice, we rarely see perfectly symmetrical distributions.</li>
<li>However, these ideal cases we learned about serve as a baseline to help us describe and interpret the distributions we see in practice.</li>

![](images/uniform_distribution.png)





```python
uniform_array = np.random.uniform(low = 20, high = 60, size = 100)
```

```python
sns.distplot(uniform_array)
```

### Skewed Distribution

<li>Not all histogram shapes are symmetrical though.</li>
<li>In a skewed distribution, we see the following:</li>
<ol>
    <li>The values pile up toward the end or the starting point of the range, making up the body of the distribution.</li>
    <li>Then the values decrease in frequency towards the opposite end, forming the tail of the distribution.</li>
</ol>


<li>If the tail points to the right, then the distribution is right skewed.</li>
<li>If the tail points to the left, then the distribution is said to be left skewed.</li>



<li>When the tail points to the left, it also points in the direction of negative numbers (on the x-axis, the numbers decrease from right to left).</li>
<li>For this reason, a left-skewed distribution is sometimes also said to have a negative skew.</li>

<li>When the tail points to the right, it also points in the direction of positive numbers. As a consequence, right-skewed distributions are also said to have a positive skew.</li>


![](images/skewed_distribution.png)

### Bargraphs

<li>A bar graph can be defined as a graphical representation of data, quantities, or numbers using bars or strips.</li>
<li>They are used to compare and contrast different types of data, frequencies, or other measures of distinct categories of data.</li>

**Elements of a Bar Graph**

<li><b>1. Bars</b></li>
<li>The rectangles drawn on the bar charts are called “bars”.</li>
<li>The bars display the number of items under particular categories.</li>

<li><b>2. Axes</b></li>
<li>There are two axes present on the graph.</li>
<li>One axis is used to represent the numerical values whereas the other represents the categorical data against which the numerical data is plotted.</li>

![](images/bar_graph.png)

<li>We often use barplots while doing univariate analysis of a categorical feature.</li>
<li>Barplots are often referred as a visual representaion of a frequency table.</li> 
<li>We can generate a bar plot using the <b>plt.bar()</b> function</li>

```python
season_df = pd.DataFrame({"season": ['Spring', "Summer", "Fall", "Winter"],
                         "No.of.students": [5, 12, 6, 10]})
season_df.head()
```

```python
plt.figure(figsize=(8, 6))
plt.bar(x=season_df['season'], height = season_df['No.of.students'])
plt.show()
```

## Customizing barplots

<li>A minimum bargraph must have categories in x-axis and height of the graph in y-axis.</li>
<li>We can add the title to the graph using <b>plt.title()</b>.</li>
<li>Similarly, we can also add x-label and y-label to the graph using <b>plt.xlabel()</b> and <b>plt.ylabel()</b> function.</li>
<li>You can also give color to each bars using <b>color</b> parameter in <b>plt.plot()</b> method.</li>
<li>Additionally, you can also supply <b>label</b> parameters to the <b>plt.plot()</b> method.</li>
<li>You can also add legend to the barplots using <b>plt.legend()</b> but you must supply label parameter in <b>plt.plot()</b> method.</li>
<li>We can also create horizontal bargraph using <b>plt.barh()</b></li>

```python
plt.figure(figsize=(8, 6))
plt.bar(x=season_df['season'], height = season_df['No.of.students'],
        label=season_df['season'], color=['red','green','orange', 'blue'])
plt.title('Arrival Of Students By Season')
plt.xlabel('Seasons')
plt.ylabel('No of Students')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(data=season_df, x='season',y='No.of.students', palette = ['red','green','orange', 'blue'])
plt.title('Arrival Of Students By Season')
plt.xlabel('Seasons')
plt.ylabel('No of Students')
plt.legend()
plt.show()
```

## Bivariate & Multivariate Analysis

<li><b>Bivariate analysis</b> refers to the analysis of two variables to determine relationships between them.</li>
<li><b>Bivariate analysis</b> explores how the dependent (“outcome”) variable depends or is explained by the independent (“explanatory”) variables.</li>
<li>If you are going to study the relationship between speeding and accident severity then such type of study comes under <b>bivariate analysis.</b></li>
<li>We can use <b>scatter plots</b>, <b>correlation heatmaps</b> between two variables and stacked bargraphs to perform bivariate analysis of the group.</li>

<li><b>Multivariate analysis</b> is one of the most useful methods to determine relationships and analyse patterns among large sets of data.</li>
<li><b>Multivariate analysis</b> is used to measure the effect of multiple independent variables on two or more dependent variables.</li>
<li>We can use pairplots, relation plots and correlation heatmaps between multiple variables to perform multivariate analysis of the group.</li>

### Scatter Plots

<li>Scatter Plot is a graph in which the values of two variables are plotted along two axes.</li>
<li>The pattern of the resulting points in a scatter plot reveals any correlation present between two data points.</li>
<li>Continuous data allows you to measure things in-depth on an infinite set. It’s commonly used for scattering analysis.</li>
<li>In matplotlib, we can use scatterplot using the <b>plt.scatter()</b> method.</li>

## Correlation

<li>When two columns are statistically related in a specific way and to a certain degree, we call this relationship <b>correlation.</b></li>

<li>There are two kinds of correlation:</li>
<ol>
    <li><b>Positive correlation</b></li>
    <li><b>Negative correlation</b></li>
</ol>

<b>1. Positive Correlation</b>
    
<li>Two positively correlated columns tend to change in the same direction. When one increases,the other tends to increaseas well.</li> 
<li>When one decreases, other tends to decrease as well.</li>
<li>On a scatter plot, two positively correlated columns show an upward trend.</li>

![](images/pos_corr.png)

<b>2. Negative Correlation</b>

<li>Two negatively correlated columns tend to change in opposite directions.</li>
<li>When one increases, the other tends to decrease, and vice versa.</li>
<li>On a scatter plot, two negatively correlated columns show a downward trend.</li>

![](images/neg_corr.png)

### Pearson's Correlation Coefficient

<li>The most popular way to measure correlation strength is by calculating the degree to which the points on a scatter plot fit on a straight line.</li>

<li>We can measure how well the points fit on a straight line by using the Pearson correlation coefficient.</li>
<li>Pearson's correlation coefficient values lie between -1.00 and +1.00.</li>
<li>When the positive correlation is perfect, the Pearson's correlation coefficient value is equal to +1.00.</li>
<li>When the negative correlation is perfect, the Pearson's correlation coefficeint value is equal to -1.00.</li>
<li>A value of 0.0 shows no correlation.</li>
<li>When we compare correlation strengths, we need to ignore the signs and only look at the absolute r values.</li>
<li>The sign only gives us the correlation's direction, not its strength.</li>

![](images/corr_types.png)

## Measuring Correlation
<li>We can measure the correlation between any numeric columns using df.corr() method.</li>
<li>But if you are interested in visualizing the correlation maps between any numeric columns, you can use seaborn heatmaps.</li>
<li><b>Heatmaps</b> are a representation of data in the form of a map or diagram in which data values are represented as colours.</li>
<li>For creating a heatmap, you need to use <b>sns.heatmap()</b> method with the following sets of parameters:</li>
<ol>
    <li>df -> a dataframe containing correlation value.</li>
    <li>annot = True or False. If you want to see the corr value inside the figure then use annot = True else False.</li>
    <li>fmt = '.1f' / '.2f' -> Its value depends on how many digits you should consider after floating value since correlation coefficient is a float value.</li>
    <li>You can specify color to the heatmap using cmap value which refers to color map.</li>
    <li>cmap = "Blues", "coolwarm", "cool", "hot", "rainbow", "flag" are some values of cmap.</li>
    <li>cbar = True or False -> Depends on whether or you want a color bar.</li>
</ol>

## Customization Of Scatter Plots

<li>For customizing a scatterplot, you can specify the colors to each of the dependent feature column.</li>
<li>We can use seaborn and matplotlib both to give beautiful customization to our scatterplots.</li>
<li>If we use matplotlib, we have to create multiple scatterplots for each of the category and give separate color and label to each category to distinguish them.</li>
<li>In matplotlib, you can give color to each scatterplot using color parameter and you can also pass label parameter to give class name.</li>
<li>Incase of seaborn, you can pass <b>hue</b> parameter to the category you want to separate the plots with.</li>
<li>Example, If you are plotting age vs bmi relationships in a scatterplot and then you want to separate it with healthy and sick category, you can specify <b>health_condition</b> column as a <b>hue</b> parameter.</li>

<code>
    sns.scatterplot(data = df, x = 'age', y = 'bmi', hue = 'health_condition', palette = ['red','green']
</code>

## Bivariate Barplots

<li>If we want to study the frequency distribution of independent column with respect to dependent column, then we use multivariate barplots.</li>
<li>We need to use pd.crosstab() method to calculate the frequency distribution of independent column with respect to dependent column.</li>
<li>Then we can create a stacked barplot to show the multivariate relation that exists between them.</li>

**syntax**

<code>
    df.plot(kind = 'bar', stacked = True, color)
</code>

## Bivariate KDE Plot

<li>A kernel density estimate (KDE) plot is a method for visualizing the distribution of observations in a dataset, analogous to a histogram.</li>
<li>Kdeplot is a Kernel Distribution Estimation Plot which depicts the probability density function of the continuous or non-parametric data variables.</li>
<li>We can plot for the univariate or multiple variables altogether.</li>
<li>For creating a bivariate KDE plot, you have to create separate dataframe belonging to a particular class or category like we did in linecharts.</li>
<li>Then you can create multiple kde plots each for each category of class.</li>
<li>You can provide separate label and separate color for each class for distinguishing each category of class.</li>
<li>This way, we can figure out the data distribution of each category of class.</li>

**Syntax**

<code>
    sns.kdeplot(df1, color, label)
    sns.kdeplot(df2, color, label)
</code>

<li>Here df1 and df2 are two separate dataframes belonging to each category of class.</li>

### Pairplots

<li>A pairplot plot a pairwise relationships in a dataset.</li>
<li>The pairplot creates a grid of axes such that each variable in data will be shared in the y-axis across a single row and in the x-axis across a single column.</li>
<li>To plot multiple pairwise bivariate distributions in a dataset, you can use the .pairplot() function.</li>

**Syntax**

<code>
    sns.pairplot(df, hue = dependent_column)
</code>

<li>We can get better 3d visualizations with plotly library.</li>
